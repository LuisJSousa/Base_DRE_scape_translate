```python
import pandas as pd
```

```python
dataset = pd.read_excel(r'excelpath.xlsx')
dataset.head()
dataset.dtypes
```

```python
correspondence = pd.read_excel(r'Correspondance_2003-2007_en.xlsx')
correspondence.head()
```

```python
correspondence.dtypes
```

```python
dic_cpvs = correspondence.to_dict("split")
```

```python
import json
print(json.dumps(dic_cpvs, sort_keys=True, indent=4))
```

```python
dic = dic_cpvs["data"]
items = {}
for line in dic:
        key, value = line[0], line[1:]
        items[key] = value
```

```python
cpvs = dataset["Cpvs"]
cpvs[149]
type(cpvs[149])
```

```python
cpvs = cpvs.astype("string")
```

```python
cpvs = cpvs.str.split("|").tolist()
print(cpvs[149])
```

```python
type(cpvs[149])
cpvs[149][1]
```

```python
aaList = []

for i in cpvs:
    j = [x.strip() for x in i]
    aaList.append(j)
```

```python
aaList[149]
```

```python
my_list = []
for k in cpvs:
    for s in k:
        if s in items:
            my_list.append(items[s])
        if s not in items: 
            my_list.append(s)
```

```python
res = [[items[i] for i in j] for j in aaList]
```

```python
for k in cpvs[149]:
    print(k)
```

```python
df = pd.DataFrame({'col':res})
print (df)
writer = pd.ExcelWriter('test.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='welcome', index=False)
writer.save()
```
